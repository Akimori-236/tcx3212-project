```python
import pandas as pd
import matplotlib.pyplot as plt
```

# LOAD DATA

```python
approval_90s = pd.read_csv("./data/Resale Flat Prices (Based on Approval Date), 1990 - 1999.csv")
approval_90s.head(3)
```

```python
approval_00s = pd.read_csv(
    "./data/Resale Flat Prices (Based on Approval Date), 2000 - Feb 2012.csv"
)
approval_00s.head(3)
```

```python
def boxplot_on_column(df: pd.DataFrame, column_name: str, vert=True):
    column_labels: list[str] = df[column_name].unique().tolist()
    column_labels.sort()
    print(column_labels)

    data_rooms_90s = [
        df[df[column_name] == label]["resale_price"] for label in column_labels
    ]
    plt.boxplot(
        data_rooms_90s,
        tick_labels=column_labels,
        vert=vert,
    )
    if vert:
        plt.ylabel("Resale Price")
    else:
        plt.xlabel("Resale Price")
```

## 90s FLAT TYPES

```python
boxplot_on_column(approval_90s, "flat_type")
plt.title("Resale prices in the 90s between different flat types")
plt.show()
```

## 2000s FLAT TYPES

```python
boxplot_on_column(approval_00s, "flat_type")
plt.title("Resale prices in the 2000s between different flat types")
plt.show()
```

```python
boxplot_on_column(approval_90s, "town", vert=False)
plt.title("Resale prices in the 90s between different towns")

plt.show()
```

```python
boxplot_on_column(approval_00s, "town", vert=False)
plt.title("Resale prices in the 2000s between different towns")
plt.show()
```
